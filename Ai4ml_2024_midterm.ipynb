### Q2-1 [5 points] Complete the empty code section below to combine the datasets into one large dataset. In your code you need to load corresponding data with column names using 'columns' information.

```python
years = range(1880, 2011)

pieces = []
columns = ['name', 'sex', 'births']

for year in years:
    path = 'yobdata/yob%d.txt' % year
    data = ''
    with open(path, 'r') as f:
      data = f.readlines()
    data_split = [x.strip().split(',') for x in data[1:]]
    frame = pd.DataFrame(data_split, columns = columns)

    frame['year'] = year
    pieces.append(frame)

# Concatenate everything into a single DataFrame
names = pd.concat(pieces, ignore_index=True)
```
